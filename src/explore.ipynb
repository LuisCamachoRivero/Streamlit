```python
import pandas as pd
```

```python
df = pd.read_csv('/workspaces/Streamlit/data/raw/StudentPerformanceFactors.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df2 = df[df['Exam_Score']<101]
df2.describe()
```

```python
from sklearn.model_selection import train_test_split

X = df2.drop(['Exam_Score'],axis=1)
y = df2['Exam_Score']

X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.8,random_state=42)

X_train.info()
```

```python
cols = X_train.columns

cols_num = df2._get_numeric_data().columns
cols_num
```

```python
cols_cat = list(set(cols)-set(cols_num))
cols_cat
```

```python
for col in cols_cat:
    df2[col+'_num'] = pd.factorize(df2[col])[0]
df3 = df2.drop(cols_cat,axis=1)
df3
```

```python
X = df3.drop(['Exam_Score'],axis=1)
y = df3['Exam_Score']

X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.8,random_state=42)

X_train.info()
```

```python
X_train.columns
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor(random_state=42)

model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
X_test.columns
```

```python
from sklearn.metrics import r2_score, root_mean_squared_error

print (f'R2 Score : {r2_score(y_test,y_pred)}')
print (f'RMSE Score : {root_mean_squared_error(y_test,y_pred)}')

```

```python
model.feature_names_in_
```

```python
model.feature_importances_
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=4)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

```

```python
X_train_pca
```

```python
model2 = DecisionTreeRegressor(random_state=42)
model2.fit(X_train_pca,y_train)

y_pred_pca = model2.predict(X_test_pca)

print (f'R2 Score : {r2_score(y_test,y_pred_pca)}')
print (f'RMSE Score : {root_mean_squared_error(y_test,y_pred_pca)}')

```

```python
print(f'Varianza PCA: {pca.explained_variance_ratio_.sum()}\nExplica {pca.explained_variance_ratio_.sum()*100:.2f}\
 % de la varianza de los datos de todas las columnas anteriores en 4 columnas')
```

```python
from pickle import dump

dump(model2,open('../models/decission_tree_classifier_42_pca.sav','wb'))
dump(pca,open('../models/pcs_4.sav','wb'))
```
